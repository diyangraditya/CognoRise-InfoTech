## Pre-processing

```python
data1 = data.drop(['car_ID'], axis=1)
data1
```

```python
symboling_mapping = {-3: 3, -2: 2, -1: 1, 0: 0, 1: -1, 2: -2, 3: -3}

data1['Symboling_Categorical'] = data1['symboling'].map(symboling_mapping)

data1
```

```python
data2 = data1.drop(['symboling'], axis=1)
data2
```

```python
test_enc = ce.OneHotEncoder(cols=['fueltype', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'fuelsystem'], handle_unknown='return_nan', 
                            return_df=True, use_cat_names=True)

data2
```

```python
data2_enc = test_enc.fit_transform(data2)
data2_enc
```

```python
test_enc1 = ce.OneHotEncoder(cols='CarName', handle_unknown='return_nan',
                            return_df=True, use_cat_names=True)

data2_enc1 = test_enc1.fit_transform(data2_enc)
data2_enc1
```

```python
#LABEL ENCODE
le_enc_aspiration = ce.OrdinalEncoder(cols=['aspiration'], return_df=True, mapping=[{'col':'aspiration',
                                                                              'mapping':{'std':0,'turbo':1}}])

data3_enc = le_enc_aspiration.fit_transform(data2_enc1)
data3_enc
```

```python
test_enc2_enginet = ce.OneHotEncoder(cols='enginetype', handle_unknown='return_nan',
                             return_df=True, use_cat_names=True)

data3_enc1 = test_enc2_enginet.fit_transform(data3_enc)
data3_enc1
```

```python
#LABEL ENCODE
le_enc_cynum = ce.OrdinalEncoder(cols=['cylindernumber'], return_df=True, mapping=[{'col':'cylindernumber',
                                                                                     'mapping':{'two':0,'three':1, 'four':2, 'five':3, 'six':4, 'eight':5, 'twelve':6}},])

data4 = le_enc_cynum.fit_transform(data3_enc1)
data4
```

```python
#SCALE
col_to_scale = ['wheelbase', 'carlength', 'carwidth','carheight', 'curbweight', 'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']

stdsc = StandardScaler()

data4[col_to_scale] = stdsc.fit_transform(data4[col_to_scale])

data4
```
